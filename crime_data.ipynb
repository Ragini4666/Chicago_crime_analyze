### FILL THE CATEGORICAL VALUES THROUGH MODE VALUE

```python
#fill NAN values =>mode function
d1['Location Description'] = d1['Location Description'].fillna(d1['Location Description'].mode())
```

```python
d1['Location Description'].isnull().sum()
```

```python
d1['Location'] =d1['Location'].fillna(d1['Location'].mode())
```

```python
d1['Location'].isnull().sum()
```

### FILL THE CONTINUOUS VALUES THROUGH MEAN VALUE  

```python
d1['Longitude'] =d1['Longitude'].fillna(d1['Longitude'].mean())
d1['Latitude'] = d1['Latitude'].fillna(d1['Latitude'].mean())
d1['Ward'] = d1['Ward'].fillna(d1['Ward'].mean())
d1['Community Area'] = d1['Community Area'].fillna(d1['Community Area'].mean())
```

```python
d1['X Coordinate'] = d1['X Coordinate'].fillna(d1['X Coordinate'].mean())
d1['Y Coordinate'] = d1['Y Coordinate'].fillna(d1['Y Coordinate'].mean())
```

```python
d1.isnull().sum()
```
